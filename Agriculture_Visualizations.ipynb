```python
# prompt: sort the agri table by the column "Greenhouse gas emissions per 1000kcal (kgCO₂eq per 1000kcal)" so that the greatest value is on the top

agri_sorted = agri.sort_values(by='Greenhouse gas emissions per 1000kcal (kgCO₂eq per 1000kcal)', ascending=False)
agri_sorted.head(100)
```

```python
# prompt: generate a new dataset called agri2 that takes the top 10 of agri_sorted

agri2 = agri_sorted.head(10)
agri2
```

```python
# try creating a data visualization (ex. graph)

#import PLT
import matplotlib.pyplot as plt
%matplotlib inline

# set size of figure
plt.rcParams['figure.figsize'] = (8.0, 6.0)
"""
#Histogram Syntax
plt.figure()
plt.hist(nrg['co2_per_unit_energy'], bins = 50, color='go')
plt.xlabel('U.S. - 2000 to present')
plt.ylabel('CO2 from Energy')
plt.title('CO2 Produced per Unit Energy in the U.S. after 2000', fontsize = 22)
plt.show()
"""
# design
csfont = {'fontname':'serif'}

# Create a bar chart
agri2 = agri_sorted.head(10)
plt.figure(figsize=(10, 6))
plt.bar(agri2['Food product'], agri2['Greenhouse gas emissions per 1000kcal (kgCO₂eq per 1000kcal)'], color = '#9BD93C')
plt.xlabel('Product', **csfont, color = 'white')
plt.ylabel('Greenhouse gas emissions per 1000kcal (kgCO₂eq per 1000kcal)', **csfont, color = 'white')
plt.title('C02 Production based on Food Product', fontsize=22, **csfont, color = 'white')
plt.xticks(agri2['Food product'], rotation=20, **csfont, color = 'white')
plt.yticks(**csfont, color = 'white')
# make bottom and left border white
plt.gca().spines['bottom'].set_color('white')
plt.gca().spines['left'].set_color('white')
plt.gca().tick_params(axis='x', colors= 'white')
plt.gca().tick_params(axis='y', colors= 'white')
#make top and right border white
plt.gca().spines['top'].set_color('white')
plt.gca().spines['right'].set_color('white')
plt.gca().tick_params(axis='x', colors= 'white')
plt.gca().tick_params(axis='y', colors= 'white')

plt.savefig('energyUSA.png', transparent=True)

plt.show()
```
